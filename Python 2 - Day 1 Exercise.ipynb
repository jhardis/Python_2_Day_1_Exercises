Exercise 2: Create a class called Rectangle with height and length attributes and has following methods: 
- perimeter() : returns a perimeter of a rectangle object
- area() : returns an area of a rectangle object

Then,
1. create two instances of Rectangle class
2. print each rectangle's perimeter and determine which rectangle has larger perimeter
3. print each rectangle's area and determine which rectangle has larger area

```python
class Rectangle :
    """Rectangle class has hight and length attributes"""
    
    def __init__ (self, height, length) :
        """Initialization method for height and length values"""
        self.height = height
        self.length = length
    
    
    def perimeter(self) :
        """Returns the perimeter of the rectangle"""
        return (self.height + self.height + self.length + self.length)
    
    
    def area(self) :
        """Returns the area of the rectangle"""
        return (self.height * self.length)


# 1. Create two instances of Rectangle class

rectangle_1 = Rectangle(2, 5)
rectangle_2 = Rectangle(3, 4)

# 2. Print each rectangle's perimeter and determine which rectangle has larger perimeter

perim_1 = rectangle_1.perimeter()
perim_2 = rectangle_2.perimeter()

print (f"The perimeter of Rectangle 1 is {perim_1}.")
print (f"The perimeter of Rectangle 2 is {perim_2}.")

if (perim_1 == perim_2) :
    print ("Their perimeters are the same.")
elif (perim_1 > perim_2) :
    print ("Rectangle 1 has the larger perimeter.")
else :
    print ("Rectangle 2 has the larger perimeter.")
print ()

# 3. Print each rectangle's area and determine which rectangle has larger area

area_1 = rectangle_1.area()
area_2 = rectangle_2.area()

print (f"The area of Rectangle 1 is {area_1}.")
print (f"The area of Rectangle 2 is {area_2}.")

if (area_1 == area_2) :
    print ("Their areas are the same.")
elif (area_1 > area_2) :
    print ("Rectangle 1 has the larger area.")
else :
    print ("Rectangle 2 has the larger area.")
```

Exercise 3: Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



```python
import math

class Triangle :
    """Triangle class has height and base attributes"""
    
    def __init__ (self, height, base) :
        """Initialization method for height and base values"""
        self.height = height
        self.base = base
    
    
    def perimeter(self) :
        """Returns the perimeter of the triangle"""
        return (self.height + self.base + math.sqrt(self.height * self.height + self.base * self.base))
    
    
    def area(self) :
        """Returns the area of the triangle"""
        return ((self.height * self.base) / 2)


# 1. Create two instances of Triangle class

triangle_1 = Triangle(6, 8)
triangle_2 = Triangle(5, 12)

# 2. Print each triangle's perimeter and determine which rectangle has larger perimeter

perim_1 = triangle_1.perimeter()
perim_2 = triangle_2.perimeter()

print (f"The perimeter of Triangle 1 is {perim_1}.")
print (f"The perimeter of Triangle 2 is {perim_2}.")

if (perim_1 == perim_2) :
    print ("Their perimeters are the same.")
elif (perim_1 > perim_2) :
    print ("Triangle 1 has the larger perimeter.")
else :
    print ("Triangle 2 has the larger perimeter.")
print ()

# 3. Print each triangle's area and determine which triangle has larger area

area_1 = triangle_1.area()
area_2 = triangle_2.area()

print (f"The area of Triangle 1 is {area_1}.")
print (f"The area of Triangle 2 is {area_2}.")

if (area_1 == area_2) :
    print ("Their areas are the same.")
elif (area_1 > area_2) :
    print ("Triangle 1 has the larger area.")
else :
    print ("Triangle 2 has the larger area.")
```
